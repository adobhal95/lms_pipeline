```python
KV_APP_CLIENT_SECRET_NAME="lmsapp-secrets"
KV_APP_TENANT_ID_NAME="lmsapp-tenantid"
KV_APP_CLIENT_NAME="lmsapp-name"
STORAGE_ACCOUNT_NAME="delakehouse"
CONTAINER_NAME = "lmsproject"
```

```python
KV_URI = "https://kvdeproject-eus.vault.azure.net/"
KV_APP_CLIENT_ID = notebookutils.credentials.getSecret(akvName=KV_URI,secret=KV_APP_CLIENT_NAME)
KV_APP_CLIENT_SECRET = notebookutils.credentials.getSecret(akvName=KV_URI,secret=KV_APP_CLIENT_SECRET_NAME)
KV_APP_TENANT_ID = notebookutils.credentials.getSecret(akvName=KV_URI,secret=KV_APP_TENANT_ID_NAME)
```

```python
# Configure Spark to use Service Principal
spark.conf.set(f"fs.azure.account.auth.type.{STORAGE_ACCOUNT_NAME}.dfs.core.windows.net", "OAuth")
spark.conf.set(f"fs.azure.account.oauth.provider.type.{STORAGE_ACCOUNT_NAME}.dfs.core.windows.net", 
               "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set(f"fs.azure.account.oauth2.client.id.{STORAGE_ACCOUNT_NAME}.dfs.core.windows.net", KV_APP_CLIENT_ID.strip())
spark.conf.set(f"fs.azure.account.oauth2.client.secret.{STORAGE_ACCOUNT_NAME}.dfs.core.windows.net", KV_APP_CLIENT_SECRET.strip())
spark.conf.set(f"fs.azure.account.oauth2.client.endpoint.{STORAGE_ACCOUNT_NAME}.dfs.core.windows.net", 
               f"https://login.microsoftonline.com/{KV_APP_TENANT_ID.strip()}/oauth2/token")
```
